```python
#Drop missing rows
sales = sales.dropna()
```

```python
#Remove the currency symbol (₹) and commas from the values in the 'actual_price' column
sales['actual_price'] = sales['actual_price'].str.replace('₹', '').str.replace(',', '')

sales['discounted_price'] =sales['discounted_price'].str.replace('₹', '').str.replace(',', '')
```

```python

# Convert the values in the 'actual_price' column to float
sales['actual_price'] = sales['actual_price'].astype(float)
# Convert the values in the 'discounted_price' column to float
sales['discounted_price'] = sales['discounted_price'].astype(float)
```

```python

# Convert 'discount_percentage' column to string data type
sales['discount_percentage'] = sales['discount_percentage'].astype(str)

# Remove percentage symbol from 'discount_percentage' column
sales['discount_percentage'] = sales['discount_percentage'].str.rstrip('%')

# Convert 'discount_percentage' to float
sales['discount_percentage'] = sales['discount_percentage'].astype(float)

#sales['rating'] = sales['rating'].astype(float)
sales.head()
```

```python
# Rename columns
sales = sales.rename(columns={'discounted_price': 'Discounted_price (₹)', 'actual_price': 'Actual_price (₹)', 'discount_percentage':'Discount_percentage (%)'})
sales.head()
```

```python
# Split the 'category' column into two separate columns
sales[['main_category', 'sub_category']] = sales['category'].str.split('|', expand=True, n=1)
```

```python
# Extract only the first part of the 'subcategory' column
sales['subcategory'] = sales['sub_category'].str.split('|').str[0]
```

```python
# Drop the 'sub_category' column
sales.drop(columns=[ 'category', 'sub_category'], inplace=True)
#sales.head()
```

```python
# Create the 'discount' column
sales['discount (₹)'] = sales['Actual_price (₹)'] - sales['Discounted_price (₹)']
#sales.head()
```

```python
# Move the 'main_category' and 'sub_category' columns to the third position
sales = sales[['product_id', 'product_name', 'main_category', 'subcategory', 'Discounted_price (₹)', 'Actual_price (₹)', 'Discount_percentage (%)', 'discount (₹)', 'rating', 'rating_count', 'about_product', 'user_id', 'user_name', 'review_id', 'review_title', 'review_content', 'img_link', 'product_link']]

# Display the updated DataFrame
sales.head()
```

SEARCH ENGINE RECOMMENDATION SYSTEM

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import string

from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

TEXT PREPROCESSING

```python

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Function to preprocess text
def preprocess_text(text):
    # Convert text to lowercase
    text = text.lower()
    
    # Remove punctuation and special characters
    text = text.translate(str.maketrans('', '', string.punctuation))
    
    # Tokenize the text into words
    tokens = word_tokenize(text)
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [token for token in tokens if token not in stop_words]
    
    # Apply stemming or lemmatization
    # Uncomment one of the following lines based on whether you want to use stemming or lemmatization
    # stemmed_tokens = [stemmer.stem(token) for token in tokens]
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
    
    # Join the tokens back into a single string
    # Uncomment one of the following lines based on whether you want to use stemming or lemmatization
    # preprocessed_text = ' '.join(stemmed_tokens)
    preprocessed_text = ' '.join(lemmatized_tokens)
    
    return preprocessed_text

# Apply preprocessing to the "about_product" column
sales['preprocessed_about_product'] = sales['about_product'].apply(preprocess_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=1000)  # You can adjust max_features as needed

# Fit and transform the preprocessed text
tfidf_matrix = tfidf_vectorizer.fit_transform(sales['preprocessed_about_product'])
```

```python
def generate_recommendations_df_with_details(search_words, tfidf_vectorizer, tfidf_matrix, df, n_recommendations=5):
    """
    Generate product recommendations with additional details and return results in a DataFrame.

    Parameters:
    - search_words (list): List of search words.
    - tfidf_vectorizer (TfidfVectorizer): Fitted TF-IDF vectorizer.
    - tfidf_matrix (sparse matrix): TF-IDF matrix of the preprocessed "about_product" text data.
    - df (DataFrame): DataFrame containing product information.
    - n_recommendations (int): Number of recommendations to generate per search word. Default is 5.

    Returns:
    - recommendations_df (DataFrame): DataFrame containing recommendations with additional details.
    """
    recommendations = []

    # Vectorize the search words
    search_word_vectors = tfidf_vectorizer.transform(search_words)

    # Calculate cosine similarity between search words and product descriptions
    similarity_scores = cosine_similarity(search_word_vectors, tfidf_matrix)

    # Get indices of top-N similar products for each search word
    top_n_indices = similarity_scores.argsort(axis=1)[:, ::-1][:, :n_recommendations]

    # Get the product details for the top-N similar products
    for i, indices in enumerate(top_n_indices):
        search_word = search_words[i]
        for idx in indices:
            product_name = sales.iloc[idx]['product_name']
            about_product = sales.iloc[idx]['about_product']
            rating = sales.iloc[idx]['rating']
            discounted_price = sales.iloc[idx]['Discounted_price (₹)']
            recommendations.append((search_word, product_name, about_product, rating, discounted_price))

    # Create DataFrame from recommendations list
    recommendations_df = pd.DataFrame(recommendations, columns=['Search_Word', 'Product_Name', 'About_Product', 'Rating', 'Discounted_Price'])
    
    return recommendations_df
```

```python
# Define the search word
search_word = "charger"

# Generate recommendations with additional details
recommendations_df = generate_recommendations_df_with_details([search_word], tfidf_vectorizer, tfidf_matrix, sales)

print("Recommendations for search word:", search_word)
print(recommendations_df)
```
